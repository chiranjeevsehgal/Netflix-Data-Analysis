This dataset consists of TV shows and movies available on Netflix as of 2019. The dataset is collected from Kaggle.com. In this notebook we are going to process and analyze the data using Numpy and Pandas libraries and then we will visualize it using matplotlib library.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('netflix_dataset.csv')
```

```python
data.drop(['director','cast','date_added'],axis=1, inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['country'].replace(np.nan, 'United States', inplace=True)
```

```python
data['rating'].replace(np.nan,'TV-MA', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data.tail()
```

```python
plt.figure(figsize=(11,4))
pd.value_counts(data['rating']).plot.bar()
plt.title('Ratings Count of Movies and Shows')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()
```

```python
last5years_df = pd.DataFrame(data['release_year'].value_counts().head())
last5years_df.reset_index(inplace=True)
last5years_df.columns = ['year', 'counts']
last5years_df.sort_values('year', ascending=False, inplace=True)
last5years_df.reset_index(inplace=True)
```

```python
plt.figure(figsize=(6,4))
plt.bar(last5years_df.year, last5years_df.counts)
plt.title('Number of releases in last 5 years')
plt.xlabel('Years')
plt.ylabel('Count')
plt.show()
```

```python
type_count_df = pd.DataFrame(data['type'].value_counts())
type_count_df.reset_index(inplace=True)
type_count_df.columns = ['title','type']
```

```python
plt.figure(figsize=(5,4))
plt.bar(type_count_df.title, type_count_df.type, width=0.2)
plt.title('Movies and Shows Count')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
most_titles_produced = data.groupby('country').count().sort_values('title',ascending=False).head()
most_titles_produced.reset_index(inplace=True)
```

```python
plt.figure(figsize=(6,4))
plt.bar(most_titles_produced.country, most_titles_produced.title)
plt.title('Top 5 countries which produced maximum titles')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.show()
```

```python
most_rated = data['rating'].value_counts()
most_rated_df = pd.DataFrame(most_rated).head()
most_rated_df.reset_index(inplace=True)
most_rated_df.columns=['title', 'rating']
```

```python
plt.figure(figsize=(6,4))
plt.bar(most_rated_df.title,most_rated_df.rating)
plt.title('5 most popular ratings')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()
```

```python
top_genres_df = pd.DataFrame(data['listed_in'].value_counts().head())
top_genres_df.reset_index(inplace=True)
top_genres_df.columns = ['title', 'listed_in']
```

```python
plt.figure(figsize=(8,4))
plt.barh(top_genres_df.title,top_genres_df.listed_in)
plt.title('5 top rated genres')
plt.xlabel('Rating')
plt.ylabel('Genres')
plt.show()
```

```python
least_genres_df = pd.DataFrame(data['listed_in'].value_counts().tail())
least_genres_df.reset_index(inplace=True)
least_genres_df.columns = ['title', 'listed_in']
```

```python
plt.figure(figsize=(8,4))
plt.barh(least_genres_df.title, least_genres_df.listed_in)
plt.title('5 least rated Genres')
plt.xlabel('Count')
plt.ylabel('Genres')
plt.show()
```
